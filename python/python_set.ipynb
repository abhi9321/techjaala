### What is a Set?
- #### unordered collection of unique elements
- #### sets do not allow duplicate values
- #### do not maintain any particular order


# Creating a Set


```python
# create set 
fruit_set = {"apple","banana","cherry"}
print(fruit_set)
```

```python
# set()
number_set = set([1,2,3,4,5])
print(number_set)
```

## Adding and Removing Elements
- #### add() method
- #### update() method
- #### remove() or discard() method

```python
# add() ele
fruit_set.add("orange")
print(fruit_set)
```

```python
# update()
fruit_set.update(["grapes","kiwi"])
print(fruit_set)
```

```python
fruit_set.remove("banana")
print(fruit_set)
```

```python
fruit_set.remove("banana")
```

```python
fruit_set.discard("orange")
```

# Set Operations
- #### Union
- #### Intersection
- #### Difference
- #### Symmetric difference



#### UNION
- #### set containing all the distinct values in both sets
  
#### set_a = {1,2,3} Set_b = {3,4,5}





```python
# Union
set_a = {1, 2, 3}
set_b = {3, 4, 5}
union_set = set_a.union(set_b)
print("Union:", union_set)
```

#### INTERSECTION
#### set_a = {1,2,3} Set_b = {3,4,5}




```python

# Intersection
intersection_set = set_a.intersection(set_b)
print("Intersection:", intersection_set)
```

#### DIFFERENCE
- #### the set of elements that are present in set_a but not in set_b.
  
#### set_a = {1,2,3} Set_b = {3,4,5}




```python
# Difference
difference_set = set_a.difference(set_b)
print("Difference:", difference_set)
```

#### SYMMETRIC DIFFERENCE
- #### the set of elements that are in either set but not in their intersection.
#### set_a = {1,2,3} Set_b = {3,4,5}



```python
# Symmetric Difference
sym_diff_set = set_a.symmetric_difference(set_b)
print("Symmetric Difference:", sym_diff_set)
```

#### set comprehension

```python
# traditional way 

set_1 = {1,2,3}
set_2 = set()
for nbr in set_1:
    set_2.add(nbr**2)
print(set_2)
```

```python
# set comprehension

sqred_set = {nbr**2 for nbr in set_1}
print(sqred_set)
```
